## 당뇨병 예측해보기

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import tensorflow as tf
```

```python
# 데이터 로드
dataset = np.loadtxt("pima-indians-diabetes.csv", delimiter=",")
```

```python
dataset
```

```python
tf.__version__
```

```python
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
print(X)
print(Y)
```

Score function 설정(keras 모델의 설정)

케라스에서는 model.add를 사용해서 레이어를 쌓을 수 있습니다.

Sequential()<br>
-> NN모델을 Sequential하게 쌓을것임.<br>
model.add(Dense(12, input_dim=8, activation='relu'))<br>
-> 8개의 노드를 가진 입력층, 12개의 노드를 가진 중간층, Dense=모든 노드를 연결 / 하나의 층(activation function을 위한)이 더 있으며, Relu activation function으로 연결

```python
# 모델의 설정
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 모델의 설정2
model = Sequential()
model.add(Dense(20, input_dim=8, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 모델 컴파일
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

# 모델 실행
X -> Y로 가는 모델을 학습하라.<br>
epochs
-> 하나의 세대(768개의 데이터)<br>
batch_size
-> 10개의 평균을 통해 Gradient Descent 갱신

```python
# 모델 실행
model.fit(X, Y, epochs=200, batch_size=10)
```

```python
X.shape
#768개의 데이터가 존재함
```

```python
# 결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, Y)[1]))
```

```python
result = model.evaluate(X, Y, verbose=0)
#verbose=0 -> 진행사항 안보임
```
